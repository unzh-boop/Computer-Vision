```python
from google.colab.patches import cv2_imshow
import cv2 
import numpy as np 
from skimage import io
```

Showing the images in raw form.

```python
img1 = cv2.imread('cat.jpg')
img2 = cv2.imread('cats_00420.jpg')

cv2_imshow(img1)
cv2_imshow(img2)
```

Applying negative filters.

```python
negative1 = cv2.bitwise_not(img1);
negative2 = cv2.bitwise_not(img2);

cv2_imshow(negative1);
cv2_imshow(negative2);
```

Log transformation with different C values

```python
log1 = (255 / np.log(1 + np.max(img1))) * (np.log(img1 + 1))
log2 = (255 / np.log(1 + np.max(img2)) ) * (np.log(img2 + 2)) 
   
log1 = np.array(log1, dtype = np.uint8) 
log2 = np.array(log2, dtype = np.uint8) 

cv2_imshow(log1)
cv2_imshow(log2)
```

Power transformation with differnt gamma values

```python
gamma11 = np.array(255*(img1/255)**10, dtype='uint8')
gamma22 = np.array(255*(img2/255)**3, dtype='uint8')

cv2_imshow(gamma11)
cv2_imshow(gamma22)
```

Sobel operator (one on x and one on y)

```python
Togray = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY )
sobelx = cv2.Sobel(cv2.GaussianBlur(Togray,(3,3),0) ,cv2.CV_8U,1,0,ksize=5)

Togray1 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY )
sobely = cv2.Sobel(cv2.GaussianBlur(Togray1,(3,3),0),cv2.CV_8U,0,1,ksize=5)

cv2_imshow(sobelx)
cv2_imshow(sobely)
```

Prewitt  (first on both, 2nd on x)


```python
prewittx = cv2.filter2D(cv2.GaussianBlur(Togray,(3,3),0), -1, np.array([[1,1,1],[0,0,0],[-1,-1,-1]]))
prewitty = cv2.filter2D(cv2.GaussianBlur(Togray,(3,3),0), -1, np.array([[-1,0,1],[-1,0,1],[-1,0,1]]))

prewittx1 = cv2.filter2D(cv2.GaussianBlur(Togray1,(3,3),0), -1, np.array([[1,1,1],[0,0,0],[-1,-1,-1]]))

cv2_imshow(prewittx + prewitty)
cv2_imshow(prewittx1)
```
